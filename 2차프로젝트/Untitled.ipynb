# 2. 데이터 수집

```python
dogData=pd.read_csv('dogComp1.csv',encoding='euc-kr')
```

```python
dogData.head()
```

# 3. 데이터 전처리

```python
#Unnamed 컬럼이름 바꾸기
dogData.rename(columns={
    dogData.columns[0]:'정답'
},inplace=True)
```

```python
dogData.info()
```

```python
dogData.head()
```

```python
#원 핫 인코딩
```

```python
onehot_columns=['친화성','털빠짐','집지키기','실내외구분','운동량','크기','거주조건','나이조건']
```

```python
for col in onehot_columns:
    dummy = pd.get_dummies(dogData[col],prefix=col)
    dogData = pd.concat([dogData,dummy],axis=1)
    dogData.drop(col,axis=1,inplace=True)
```

```python
dogData
```

```python
#체고와 체중은 크기로 구분할 수 있기때문에 컬럼을 삭제하겠다.
dogData.drop('체고',axis=1,inplace=True)
dogData.drop('체중',axis=1,inplace=True)
```

```python
dogData.info()
```

```python
dogData.loc[:,'친화성_h':]
```

```python
dogData.loc[:,'친화성_h']
```

```python
#문제와 정답으로 분리
y=dogData['정답']
X=dogData.loc[:,'친화성_h':]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                test_size=0.3,
                                                random_state=3)
```

```python
logi = LogisticRegression()
logi.fit(X_train,y_train)
print("train score: ",logi.score(X_train,y_train))
print("test score: ",logi.score(X_test,y_test))
```
